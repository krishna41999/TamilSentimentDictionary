```python
def pmi(word1, word2, unigram_freq, bigram_freq):
    unigram_freq_values=0
    bigram_freq_values=0
    unigram_freq_word1=0
    unigram_freq_word2=0
    j=0
    for i in unigram_freq:
        if unigram_freq[j][0][0]==word1:
            unigram_freq_word1=unigram_freq[j][1]
        if unigram_freq[j][0][0]==word2:
            unigram_freq_word2=unigram_freq[j][1] 
        j=j+1
    i=0
    for k in bigram_freq:
        if bigram_freq[i][0][0]==word1 and bigram_freq[i][0][1]==word2:
            bigram_freq1=bigram_freq[i][1]
        i=i+1
    prob_word1 = unigram_freq_word1 / len(fdistu)
    prob_word2 = unigram_freq_word2 / len(fdistu)
    prob_word1_word2 = bigram_freq1 / len(fdistb)
    a=prob_word1_word2 /(prob_word1*prob_word2)
    return round(math.log(a,10),2)
```
